The Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import pickle
```

Load Data

```python
from google.colab import drive

#mount drive
drive.mount('/content/drive')

#membaca file csv
df = pd.read_csv('/content/drive/your path directory/namefile.csv')

df.head()
```

Data Preprocessing

```python
#pengecekan data null
df.isnull().sum()
```

```python
#encoding untuk mengubah kolom menjadi nilai numerik

# Membuat encoder untuk setiap kolom kategorikal
encoder_gender = LabelEncoder().fit(df['Gender'])
encoder_vehicle_age = LabelEncoder().fit(df['Vehicle_Age'])
encoder_vehicle_damage = LabelEncoder().fit(df['Vehicle_Damage'])

# Transformasi data pelatihan dengan encoder yang telah dilatih
df['Gender'] = encoder_gender.transform(df['Gender'])
df['Vehicle_Age'] = encoder_vehicle_age.transform(df['Vehicle_Age'])
df['Vehicle_Damage'] = encoder_vehicle_damage.transform(df['Vehicle_Damage'])
```

```python
print("Gender categories in training:", encoder_gender.classes_)
print("Vehicle_Age categories in training:", encoder_vehicle_age.classes_)
print("Vehicle_Damage categories in training:", encoder_vehicle_damage.classes_)
```

```python
# Simpan encoder ke dalam file pickle
with open('/content/drive/your path directory/encoder_name1.pkl', 'wb') as f:
    pickle.dump(encoder_gender, f)

with open('/content/drive/your path directory/encoder_name2.pkl', 'wb') as f:
    pickle.dump(encoder_vehicle_age, f)

with open('/content/drive/your path directory/encoder_name2.pkl', 'wb') as f:
    pickle.dump(encoder_vehicle_damage, f)
```

Split Data

memisahkan dataset menjadi fitur x dan targer y dan membagi data menjadi data latih dan data uji

```python
#feature and target
x = df.drop(columns=['id','Response'])
y = df['Response']

#data latih dan data uji
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

Train Model

```python
#inisalisasi model RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)

#latih model dengan data latih
model.fit(x_train, y_train)
```

```python
df.head()
```

Evaluate Model

```python
#prediksi pada data uji
y_pred = model.predict(x_test)

#evaluasi model
print(classification_report(y_test, y_pred))
```

Save Model

simpan model dengan file .pkl

```python
#simpan model dengan file .pkl
with open('/content/drive/your path directory/ML_model_name.pkl', 'wb') as file:
    pickle.dump(model, file)
```
